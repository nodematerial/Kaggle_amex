```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('input/train_data.parquet')
```

```python
num_log = np.unique(df['customer_ID'].value_counts(), return_counts= True)[1]
num_log
```

```python
sns.countplot(df['customer_ID'].value_counts())
```

```python
fig, ax = plt.subplots(figsize=(75,15))
ax = sns.countplot(x="S_2", data = df.sort_values('S_2'))
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```

```python
feature = df.groupby('target').mean()
```

```python
false_feature = feature.iloc[0]
true_feature  = feature.iloc[1]
```

```python
fig, ax = plt.subplots(figsize=(75,15))
sns.barplot(x = feature.columns, y = false_feature)
```

```python
fig, ax = plt.subplots(figsize=(75,15))
sns.barplot(x = feature.columns, y = true_feature)
```

```python
fig, ax = plt.subplots(figsize=(45,10))
sns.barplot(x = feature.columns, y = true_feature/false_feature)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```
